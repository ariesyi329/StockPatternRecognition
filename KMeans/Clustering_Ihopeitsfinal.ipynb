### KMeans Algorithms

```python
def kmeas_fun(X, num_clusters):
    kmeans = KMeans(init='k-means++', n_clusters=num_clusters, n_init=10)
    kmeans.fit(X)
    result = kmeans.predict(X)
    center = (kmeans.cluster_centers_)
    distance = (kmeans.inertia_)
    return result, center, distance
```

```python
def group_result(result,num_clusters):
    d = {}
    for j in range(num_clusters):
        d[j] = []
    for i in range(len(result)):
        for j in range(num_clusters):
            if result[i] == j:
                d[j].append(i)
    return d
```

```python
def choose_cluster_num(X,fun):
    kmeans_dist = []
    for i in [1,2,4,8,16,32,64,128,256]:
        r,c,d = fun(X, i)
        kmeans_dist.append(d)
    return kmeans_dist
```

```python
def find_local_optimal(X, fun, num_of_clusters):
    min_d = 10000000000000000000
    result_r = {}
    for i in range(10):
        r,c,d = fun(X, num_of_clusters)
        if d < min_d:
            min_d = d
            result_r = group_result(r,num_of_clusters)
            opti_r = r
            opti_c = c
    return min_d, result_r, opti_r, opti_c
```

```python
def check_result(num_of_clusters, train_result, vali_result):
    check = {}
    for i in range(num_of_clusters):
        check[i] = {}
        for j in range(num_of_clusters):
            check[i][j] = 0
            
    for k in range(num_of_clusters):
        for i in train_result[k]:
            for j in range(num_of_clusters):
                if i in vali_result[j]:
                    check[k][j] += 1
                    
    accu = {}
    for p in range(num_of_clusters):
        s = sum(check[p].values())
        m = max(check[p].values())
        try:
            ac = float(m)/s
            accu[p] = ac
        except:
            pass
    return accu, check
```

```python
def check_balance(result_dict):
    count = []
    for i in range(8):
        count.append(len(result_dict[i]))
    std = np.std(count)
    m = np.mean(count)
    balance = float(std)/m
    return balance, count
    
    
```
