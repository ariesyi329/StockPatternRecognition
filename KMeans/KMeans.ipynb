```python
def kmeas_fun(X, num_clusters):
    kmeans = KMeans(init='k-means++', n_clusters=num_clusters, n_init=10)
    kmeans.fit(X)
    result = kmeans.predict(X)
    center = (kmeans.cluster_centers_)
    return result, center
    
```

```python
def distance_between_clusters(X,d_result,center,num_clusters):
    
    s = {}
    for j in range(num_clusters):
        s[j] = []
        
    cov = np.cov(X.T)
    inv = np.linalg.inv(cov)
    
    li_clusters = range(num_clusters)
    for j in li_clusters:
        for i in range(len(d_result[j])):
            distances = []
            for k in li_clusters:
                #if k != j:
                    
                #distances.append(mahalanobis(X[d_result[j][i]],center[k],inv))
                distances.append(np.linalg.norm(X[d_result[j][i]]-center[k]))
            s[j].append(distances)
    return s
```

```python
def distance_within_cluster(X,d_result,center,num_clusters):
    
    s = {}
    for j in range(num_clusters):
        s[j] = []
        
    cov = np.cov(X.T)
    inv = np.linalg.inv(cov)
    
    for j in range(num_clusters):
        for i in range(len(d_result[j])):
            #s[j].append(mahalanobis(X[d_result[j][i]],center[j],inv))
            s[j].append(np.linalg.norm(X[d_result[j][i]]-center[j]))
    
    return s
```
